```python
from pandas import *
from numpy import *
rurales=read_csv('rurales_12_02_23.csv').drop('Unnamed: 0',axis=1)
```

```python
for i in rurales.loc[rurales.otros_trab>0].index:
    rurales.tipo_hogar[i]='UFR'
```

```python
rurales['IngresoTrab']=rurales['trabajo']+rurales['otros_trab']
```

```python
rurales['IngresoCorr']=rurales['IngresoTrab']+rurales['agrope']+rurales['noagrop']+rurales['rentas']+rurales['transfer']+rurales['otros_ing']
```

```python
rurales['IngresoMensual']=0.3*rurales['IngresoCorr'].round(2)
```

```python
rurales.drop('ingreso mensual',axis=1)
```

```python
rurales.IngresoMensual=rurales.IngresoMensual.round(2)
```

```python
rurales.IngresoMensual
```

```python
rurales.to_csv('rurales_22_02_23.csv', index=False)
```

```python
rurales.filter(['IngresoCorr','IngresoTrab','agrope','noagrop','rentas','transfer','otros_ing','tipo_hogar','CUmun2','Region']).pivot_table(index=['tipo_hogar','CUmun2'],aggfunc=mean).round(2)
```

```python
THR1=rurales.filter(['factor','Region','tipo_hogar']).pivot_table(index=['Region'],columns=['tipo_hogar'],aggfunc=sum,margins=True,margins_name='Total')
THR2=rurales.filter(['folioviv','Region','tipo_hogar']).pivot_table(index=['Region'],columns=['tipo_hogar'],aggfunc='count',margins=True,margins_name='Total')
```

```python
THR1=THR1.rename(columns={'factor':'Hogares'})
```

```python
THR2=THR2.droplevel(0,axis=1)
```

```python
THR1=THR1.droplevel(0,axis=1)
```

```python
for cn in THR1.columns:
    THR1[cn + ' %']=100*THR1[cn]/THR1['Total']
```
