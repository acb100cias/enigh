```python
import bamboolib as bam
from pandas import *
from numpy import *
data=read_csv('rurales_cuMun_26_8_2021.csv').drop('Unnamed: 0',axis=1)
```

```python
data['Ingreso mensual']=data['factor']*(data['ing_cor']/3.)
data['ingreso mensual']=data['ing_cor']/3.
```

```python
data.to_csv('rurales_cuMun_22_3_2022.csv',index=False)
```

```python
data.filter(['factor','Region','tipo_hogar','Ingreso mensual']).pivot_table(index='Region',columns='tipo_hogar',aggfunc=[sum,mean],margins=True,margins_name='Total').rename(columns={'factor':'Hogares'}).astype(int)
```

```python
T11=data.filter(['factor','Region','tipo_hogar','ingreso mensual']).pivot_table(index='Region',columns='tipo_hogar',aggfunc=[sum],margins=True,margins_name='Total').rename(columns={'factor':'Hogares'}).astype(float).round(2)
```

```python
T11=T11.droplevel(level=0,axis=1)
```

```python

T11.to_excel('Ingreso corriente mensual por tipo de hogar por region.xls')
```

```python
T12=data.filter(['factor','Region','tipo_hogar','ingreso mensual']).pivot_table(index='Region',columns='tipo_hogar',aggfunc=[sum,mean],margins=True,margins_name='Total').rename(columns={'factor':'Hogares'}).astype(float).round(2)
```

```python
T12=T12.drop([('mean','Hogares'),('sum','ingreso mensual')],axis=1)
```

```python
T12=T12.droplevel(level=0,axis=1)
```

```python
T12.to_excel('Ingreso corriente mensual promedio por tipo de hogar por region.xls')
```

```python
T13=data.filter(['factor','Region','tipo_hogar','ingreso mensual']).pivot_table(index='Region',columns='tipo_hogar',aggfunc=[sum,],margins=True,margins_name='Total').rename(columns={'factor':'Hogares'}).astype(float).round(2)
```

```python
T13=data.groupby('Region')['ingreso mensual'].describe().join(data.groupby('Region')['factor'].sum()).rename(columns={'factor':'Hogares','count':'muestra'})
```

```python
T13=T13.append(DataFrame(T13.sum()).T.rename(index={0:'Total'}))
```

```python
T13=T13.round(2)
```

```python
T13.to_excel('tabla_13_descriptivos_ingreso_mensual.xls')
```
